```python
# this is just to set up headers for the csv file.

import csv
import datetime

car_features = {'title': car.find(class_="hidden-sm hidden-md hidden-lg text-left").string, 
                        'VIN': 'None', 
                        'Mileage': 'None', 
                        'Color': 'None', 
                        'Asking Price': list_of_cars[0].find(class_="price-value").string, 
                        'Car Blurb': 'None', 
                        'Car Detail Page': car.find("input", {"id": "link_1"})['value'], 
                        'Window Sticker': 'None',
                        'Date': datetime.datetime.now()}

with open('bellevue_leafs.csv', 'w') as f:
    fieldnames = car_features.keys()
    writer = csv.DictWriter(f, fieldnames=fieldnames)
    writer.writeheader()

```

```python
# now this will process the pulled data and put it into the csv.

def pull_details_bellevue_nissan(car_list):
    for car in car_list:
        car_features = {'title': car.find(class_="hidden-sm hidden-md hidden-lg text-left").string, 
                        'VIN': 'None', 
                        'Mileage': 'None', 
                        'Color': 'None', 
                        'Asking Price': "None", 
                        'Car Blurb': 'None', 
                        'Car Detail Page': 'None', 
                        'Window Sticker': 'None',
                        'Date': datetime.datetime.now()}
        
        for i in range(0, len(car.find_all(class_="field-label"))):
            if car.find_all(class_="field-label")[i].string == 'VIN:':
                car_features['VIN'] = car.find_all(class_="field-value")[i].string
            elif car.find_all(class_="field-label")[i].string == 'Mileage:':
                car_features['Mileage'] = car.find_all(class_="field-value")[i].string
            elif car.find_all(class_="field-label")[i].string == 'Exterior Color:':
                car_features['Color'] = car.find_all(class_="field-value")[i].string
        
        try:
            car_features['Car Detail Page'] = car.find("input", {"id": re.compile("^link_")})['value']
        except Exception:
            pass
        
        try:
            car_features['Car Blurb'] = car.find(class_="start-text").string
        except Exception:
            pass
        
        try:
            car_features['Asking Price'] = car.find(class_="price-value").string
        except Exception:
            pass
                
        if car_features['VIN']:
            car_features['Window Sticker'] = 'http://cpo.nissanusa.com/nna/?' + car_features['VIN']
        
        for_csv = list(car_features.values())
        
        with open('bellevue_leafs.csv', 'a') as file:
            writer = csv.writer(file)
            writer.writerow(for_csv)

            
pull_details_bellevue_nissan(list_of_cars)
```
