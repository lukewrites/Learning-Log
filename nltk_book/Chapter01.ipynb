## 1.   Computing with Language: Texts and Words

### 1.3 Searching text

We can use the `.concordance("<term>")` to search a text

```python
from nltk.book import *

text1.concordance("monstrous")
```

```python
text3.concordance('begat')
```

The `.concordance` method shows us words in context, `.similar` shows words that appear in a similar context.

```python
text1.similar('monstrous')
```

```python
text2.similar('monstrous')
```

The `.common_contexts([x, y,...])` method shows contexts shared by 2+ words.

```python
text2.common_contexts(['monstrous', 'very', 'amazingly'])
```

The _location_ of a word in a text, counted from the first word), can be shown with a **_dispersion plot_**:

```python
text4.dispersion_plot(['citizens', 'democracy', 'monarch', 'king', 'freedom'])
```

### 1.4 Counting Vocabulary

Typical operations like `len(text)`, `sorted()`, `set()` all work on a `text` object.

```python
len(text3)
```

```python
sorted(set(text3))
```

This gives us all the **tokens** (_sequences of characters_) in `text3`, `sort`ed in alphabetic order. How many unique tokens (**_word types_**) are there in `text3`?

```python
len(set(text3))
```

Holy moley!

To calculate **_lexical richness_** (the number of distinct words) of the text, we divide:

```python
len(set(text3)) / len(text3)
```

This means that each word is repeated about 16 times:

```python
1 / (len(set(text3)) / len(text3))
```

We can count how many times a particular word is used in a text and then calculate what percentage of the text is represents:

```python
text3.count('smote')
```

```python
(text4.count('a') / len(text4)) * 100
```

```python
text5.count('lol') / len(text5)
```

## 3. Computing with Language: Simple Statistics

### 3.1 Frequency Distribution

We can use _frequency distribution_ to give the see how often each item appears in a text.

This can be done with the `FreqDist` class:

```python
fdist1 = FreqDist(text1)
print(fdist1)
```

There are various class methods for `FreqDist`, including `.most_common()`:

```python
fdist1.most_common(25)
```

This can also be represented as a plot:

```python
fdist1.plot(25, cumulative=False)
```

I need to find out what this `cumulative` setting means...it seems counterintuitive.

Possibly more interesting are **_hapaxes_**, words that occur only once:

```python
fdist1.hapaxes()
```

Note that phrases ('Herman Melville') occur as lists. Interesting...how does that work?

```python
len(fdist1.hapaxes())
```

Wow, that seems like a lot of unique words_Moby Dick_ surely must have a high _lexical richness_, right?

```python
len(fdist1.hapaxes())/len(text1)
```

Wow...almost 3% of the text is unique words!

WOuld we find more useful words if we only look at long words? We can use a list comprehension to do this.

```python
V = set(text1)
long_words = [word for word in V if len(word) > 15]

sorted(long_words)
```

While long words are often hapaxes, perhaps we should look for frequently occuring long words:

```python
fdist5 = FreqDist(text5)
sorted(w for w in set(text5) if len(w) > 7 and fdist5[w] > 7)
```

### 3.3 Collocations and Bigrams

**_collocation_**: a sequence of words that occur together unusually often. In the `NLTK`, these seem to be **_bigrams_** or word pairs.  
The `bigrams()` function extracts a list of word pairs from a list of words, and `.collocations()` method pulls out bigrams from a `text`:

```python
from nltk import bigrams

list(bigrams(['The', 'dunces', 'are', 'in', 'confederacy', 'against', 'him']))
```

```python
text5.collocations()
```
