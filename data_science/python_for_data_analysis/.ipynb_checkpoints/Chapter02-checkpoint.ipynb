### 1. usa.gov data



```python
# load data from a text file
path = 'pydata-book/ch02/usagov_bitly_data2012-03-16-1331923249.txt'

# now let's take a look at it
open(path).readline()
```

```python
# let's make that prettier by using the `json` module.
import json

records = [json.loads(line) for line in open(path, 'rb')]
records[0]
```

That's much prettier! Now let's do something with the data.

```python
time_zones = [rec['tz'] for rec in records if 'tz' in rec]
# a lovely list comprehension that only pulls the tz's from recs that include one.

time_zones[:10]
```

So that's cool, but we need to do some **data cleaning** since some of the `tz`s are just `''`.

```python
from collections import defaultdict

def get_counts2(sequence):
    counts = defaultdict(int)
    for x in sequence:
        counts[x] += 1
    return counts

def get_counts(sequence):
    counts = {}
    for x in sequence:
        if x in counts:
            counts[x] += 1
        else:
            counts[x] = 1
    return counts

print("get_counts(time_zones):", get_counts(time_zones), '\n')

print("get_counts2(time_zones):", get_counts2(time_zones))
```

```python
counts = get_counts(time_zones)
counts['America/New_York']
```

```python
# for the top 10 time zones:

def top_counts(count_dict, n=10):
    value_key_pairs = [(count, tz) for tz, count in count_dict.items()]
    # that creates a list of tuples for all k:v pairs in the dict
    value_key_pairs.sort()
    # sort them, then slize the last 10 to return
    return value_key_pairs[-n:]

top_counts(counts)
```

```python
# but why not use collections.Counter?

from collections import Counter
counts = Counter(time_zones)
counts.most_common(10)
```

```python
# Let's look at what browser users were on.

results = Series([x.split()[0] for x in frame.a.dropna()])
results[:5]
```

```python
results.value_counts()[:8]
```

```python
# now, how about looking at top time zones for Windows and non-Windows users

import numpy as np  # the book was missing this part?

# clean data; this creates a new frame from a list of those `a` elements that are not Null
cframe = frame[frame.a.notnull()]

# select operating system
operating_system = np.where(cframe['a'].str.contains('Windows'), 'Windows', 'Not Windows')
operating_system[:5]
```

```python
by_tz_os = cframe.groupby(['tz', operating_system])

agg_counts = by_tz_os.size().unstack().fillna(0)
agg_counts[:10]
```

```python
# now an indirect index array from the row counts in agg_counts:

indexer = agg_counts.sum(1).argsort()
indexer[:10]
```

```python
# use `take` to select rows in order
count_subset = agg_counts.take(indexer)[-10:]

count_subset
```

```python
# to plot in a stacked bar plot,
# which is to say we stack Not Windows on top of Windows:

count_subset.plot(kind='barh', stacked=True)
```

```python
# now we will normalize the rows to see relative percentages of Windows users

normed_subset = count_subset.div(count_subset.sum(1), axis=0)
normed_subset.plot(kind='barh', stacked=True)
```

## MovieLens 1M Data Set

Movie ratings data from 90s and 00s.

```python
unames = ['user_id', 'gender', 'age', 'occupation', 'zip']
users = pd.read_table('pydata-book/ch02/movielens/users.dat', sep='::', header=None, names=unames, engine='python')
```

```python
users[:5]
```

```python
rnames = ['user_id', 'movie_id', 'rating', 'timestamp']
ratings = pd.read_table('pydata-book/ch02/movielens/ratings.dat', sep='::', header=None, names=rnames, engine='python')

ratings[:5]
```

```python
mnames = ['movie_id', 'title', 'genre']
movies = pd.read_table('pydata-book/ch02/movielens/movies.dat', sep='::', header=None, names=mnames, engine='python')
movies[:5]
```

```python
# merge the three data sets together.
# first, merge ratings with users, then merge that with movies

data = pd.merge(pd.merge(ratings, users), movies)
```

```python
data
```

```python
data.ix[0]
```

```python
# using a pivot table

mean_ratings = data.pivot_table('rating', 'title', 'gender')
# had to fix the book code a little bit to get that to work.

mean_ratings[:10]
## interesting that the index slice is for total number of ratings, e.g. 2 per movie.
```

```python
# let's only look at movies with 250+ ratings.
ratings_by_title = data.groupby('title').size()

active_titles = ratings_by_title.index[ratings_by_title >= 250]
active_titles
```

```python
mean_ratings = mean_ratings.ix[active_titles]
mean_ratings
```

```python
# now to sort our results

top_female_ratings = mean_ratings.sort_index(by='F', ascending=False)

top_female_ratings[:10]
```

#### measuing disagreement

```python
mean_ratings['diff'] = mean_ratings['M'] - mean_ratings['F']

sorted_by_diff = mean_ratings.sort_index(by='diff')

sorted_by_diff[:5]
```

```python
# now for movies that men like more than women.
# find them by going from the bottom of the list.
sorted_by_diff[::-1][:15]
```

#### look at standard deviation to find most disagreement

```python
rating_std_by_title = data.groupby('title')['rating'].std()
rating_std_by_title = rating_std_by_title.ix[active_titles]
rating_std_by_title.sort_values(ascending=False)[:10]  # had to fix the book's code here.
```

#### pandas.concat
to put together data from different files.

```python
years = range(1880, 2011)

pieces = []
columns = ['name', 'sex', 'births']

for year in years:
    path = 'pydata-book/ch02/names/yob%d.txt' % year
    frame = pd.read_csv(path, names=columns)
    
    frame['year'] = year
    pieces.append(frame)
    

# concatenate everything into a single DataFrame
names = pd.concat(pieces, ignore_index=True)
```

```python
names
```

```python
total_births = names.pivot_table('births', 'year', 'sex', aggfunc=sum)

total_births.tail()
```

```python
total_births.plot(title='Total births by sex and year')
```

#### .apply(`function`) to do something to data
in this case, it's add a new column!

```python
def add_prop(group):
    births = group.births.astype(float)
    group['prop'] = births / births.sum()
    return group


names = names.groupby(['year', 'sex']).apply(add_prop)

names
```

```python
def get_top1000(group):
    return group.sort_index(by='births', ascending=False)[:1000]

grouped = names.groupby(['year', 'sex'])
top1000 = grouped.apply(get_top1000)
top1000.index = np.arange(len(top1000))

top1000
```

```python
boys = top1000[top1000.sex == 'M']
girls = top1000[top1000.sex == 'F']
```

#### use a time series to plot

```python
total_births = top1000.pivot_table('births', 'year', 'name')
total_births.info()
```

```python
subset = total_births[['John', 'Harry', 'Mary', 'Marilyn']]
subset.plot(subplots=True, figsize=(12,10), grid=True, title='Number of births per year')
```

```python
table = top1000.pivot_table('prop', 'year', 'sex', aggfunc=sum)
table.plot(title='Sum of the table1000.prop by year and sex',
           yticks=np.linspace(0, 1.2, 13),
           xticks=range(1880, 2020, 10))
```

This shows that fewer girls than boys are given names in the top 1000 names in 1880.

```python
# let's try to calculate the number of boy names in 2010

df = boys[boys.year == 2010]
df
```

```python
# let's find the half-way point of this list...

prop_cumsum = df.sort_index(by='prop', ascending=False).prop.cumsum()

prop_cumsum[:10]
```

```python
# we are taking the cumulative sum of `prop` and then searching through `prop` sorted to see
# where we would have to put 0.5 [the half-way point] in the list to keep it sorted.
prop_cumsum.values.searchsorted(0.5)
```

```python
df = boys[boys.year == 1900]
in1900 = df.sort_index(by='prop', ascending=False).prop.cumsum()
in1900.values.searchsorted(0.5) + 1  # add 1 since we index from 0.
```

Let's write a function to apply this operation to each year/sex combo!

```python
def get_quantile_count(group, q=0.5):
    group = group.sort_index(by='prop', ascending=False)
    return group.prop.cumsum().values.searchsorted(q) + 1

diversity = top1000.groupby(['year', 'sex']).apply(get_quantile_count)
diversity = diversity.unstack('sex')
diversity.head()
```

```python
diversity.plot(title="Number of popular names in top 50%")
```

### The 'Last Letter' Revolution

Let's aggregate the births by year, sex, and final letter of name:

```python
# extract last letter from name column
get_last_letter = lambda x: x[-1]
last_letters = names.name.map(get_last_letter)
last_letters.name = 'last_letter'

table = names.pivot_table('births', last_letters, ['sex', 'year'], aggfunc=sum)

subtable = table.reindex(columns=[1910, 1960, 2010], level='year')
subtable.head()
```

```python
# now we normalize the table to make a table of proportions of total births for each sex
# ending in each letter:

subtable.sum()
```

```python
letter_prop = subtable / subtable.sum().astype(float)
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 1, figsize=(10, 8))
letter_prop['M'].plot(kind='bar', rot=0, ax=axes[0], title='Male')
letter_prop['F'].plot(kind='bar', rot=0, ax=axes[1], title='Female', legend=False)
```
